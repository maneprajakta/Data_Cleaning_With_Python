<h1>Determine the Number of Nan entities present in every column</h1>
It is important to first know whether columns are nan or not .

```python
df_na = df.isna().sum()   #this will sum nan values over every individual column in dataset
df_na        
```

<p>As we can see column "Agency Project ID" has 1181 nan values almost half of dataset enteries.</p>
we will look now how to reduce the nan value's as it's not always nice to delete all nan cell from datset.

<h5>If we look at dataset it can be foun that column Unique Investment Identifier contains  "Total"  as entry in it which needs to be removed as it's wrong .</h5>

```python
df["Unique Investment Identifier"].unique()
```

```python

df = df[df["Unique Investment Identifier"] != "Total"]  #remove all rows for which Unique Investment
                                                         #Identifier has Total value
df.shape
```

```python
df_na = df.isna().sum()
df_na

#we will again check for nan values count for each column 
```

<h4>This is intresting removing rows which had "total" value for column Unique Investment Identifier has reduced na to zero for many columns.</h4>
